### T1 — Inspect Raw Data
Load `samplesuperstore.csv` and run the four-line health check: shape, dtypes, null counts, and descriptive stats for the four numeric columns.

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\jarir\Desktop\OJT BCA\project 4.1\data\samplesuperstore.csv")

print("Shape:", df.shape)
print()
print(df.dtypes)
print()
print(df.isnull().sum())
print()
print(df[["Sales", "Quantity", "Discount", "Profit"]].describe())
```

### T2 — Check Duplicates
Count fully duplicate rows, then display every row involved in a duplication (both copies) filtered by Order ID.

```python
print("Duplicate rows:", df.duplicated().sum())

dupes = df[df.duplicated(subset=["Order ID"], keep=False)]
print("Rows involved in duplication:", dupes.shape)
dupes.head()
```

### T3 — Date Arithmetic
Convert Order Date and Ship Date from text to datetime, then create a `Shipping Days` column and print average shipping days by Region.

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"]  = pd.to_datetime(df["Ship Date"])

df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

print(df.groupby("Region")["Shipping Days"].mean().round(1))
```

### T4 — Profit Margin
Add a `Profit Margin %` column and print the average margin per Sub-Category sorted from lowest to highest.

```python
df["Profit Margin %"] = (df["Profit"] / df["Sales"] * 100).round(2)

print(df.groupby("Sub-Category")["Profit Margin %"]
        .mean()
        .sort_values()
        .head(5)
        .round(1))
```

## End-of-Day Project — Full Preprocessing Pipeline

## Step 1 — Imports and File Load
Import pandas and numpy, set the file path, and load the raw CSV into a DataFrame named `df`.

```python
import pandas as pd
import numpy as np

RAW_PATH = r"C:\Users\jarir\Desktop\OJT BCA\project 4.1\data\samplesuperstore.csv"

df = pd.read_csv(RAW_PATH)
print("File loaded successfully.")
print("Shape:", df.shape)
```

## Step 2 — Data Health Inspection
Run the four-line checklist to understand the dataset size, column types, missing values, and a preview of raw data before making any changes.

**Observations to fill in after running:**
- `Order Date` and `Ship Date` are dtype `object`, not datetime — need conversion
- No missing values found in any column
- Sales and Profit have wide ranges — worth inspecting for outliers

```python
print("=== Shape ===")
print(df.shape)

print("\n=== Data Types ===")
print(df.dtypes)

print("\n=== Missing Values ===")
print(df.isnull().sum())

print("\n=== First 3 Rows ===")
df.head(3)
```

## Step 3 — Check and Remove Duplicates
Count fully duplicate rows, display them for inspection, then drop them. Print the shape before and after so we can see how many rows were removed.

```python
print("Shape BEFORE removing duplicates:", df.shape)
print("Duplicate rows found:", df.duplicated().sum())

# Display all rows involved in duplication (both copies)
dupes = df[df.duplicated(subset=["Order ID"], keep=False)]
print("\nRows involved in duplication:")
print(dupes.shape)
dupes.head(10)
```

```python
# Drop fully duplicate rows
df.drop_duplicates(inplace=True)

print("Shape AFTER removing duplicates:", df.shape)
```

## Step 4 — Convert Date Columns
Convert `Order Date` and `Ship Date` from text (object) to datetime64 so we can do date arithmetic in Step 5. Confirm the conversion by printing dtypes.

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"]  = pd.to_datetime(df["Ship Date"])

# Confirm both columns are now datetime64
print(df[["Order Date", "Ship Date"]].dtypes)
```

## Step 5 — Engineer: Shipping Days
Subtract Order Date from Ship Date and extract the integer number of days using `.dt.days`. This measures how long each delivery actually took.

```python
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

print("Shipping Days — min / max / mean:")
print(f"  Min : {df['Shipping Days'].min()} days")
print(f"  Max : {df['Shipping Days'].max()} days")
print(f"  Mean: {df['Shipping Days'].mean().round(1)} days")
```

## Step 6 — Engineer: Order Year
Extract the year from `Order Date` using `.dt.year` and store it in a new column. This is used for year-over-year comparison charts in Phase 3.

```python
df["Order Year"] = df["Order Date"].dt.year

print("Unique years in the dataset:")
print(sorted(df["Order Year"].unique()))
```

## Step 7 — Engineer: Profit Margin %
Calculate profit as a percentage of sales. Negative values mean the company lost money on that sale — usually caused by high discounts. This column feeds directly into Phase 3 profitability charts.

```python
df["Profit Margin %"] = (df["Profit"] / df["Sales"] * 100).round(2)

print("Bottom 5 Sub-Categories by average Profit Margin %:")
print(
    df.groupby("Sub-Category")["Profit Margin %"]
      .mean()
      .sort_values()
      .head(5)
      .round(1)
)
```

## Step 8 — Rename Columns for Clean Code
Rename columns that have spaces or inconsistent capitalisation so they are easier to reference in Phase 3 code without needing quotes around every column name.

```python
df = df.rename(columns={
    "Sub-Category":  "sub_category",
    "Ship Mode":     "ship_mode",
    "Customer Name": "customer_name",
})

print("Updated column names:")
print(df.columns.tolist())
```

## Step 9 — Final Inspection and Export
Print a summary of the cleaned DataFrame — final shape, all column names, and stats on the three new engineered columns. Then save to `superstore_clean.csv`, reload, and confirm the shape matches.

```python
print("=== Final Shape ===")
print(df.shape)

print("\n=== All Columns ===")
print(df.columns.tolist())

print("\n=== Engineered Columns Summary ===")
print(df[["Shipping Days", "Order Year", "Profit Margin %"]].describe().round(2))
```

```python
# Export
df.to_csv(r"C:\Users\jarir\Desktop\OJT BCA\project 4.1\data\superstore_clean.csv", index=False)

# Verify
check = pd.read_csv(r"C:\Users\jarir\Desktop\OJT BCA\project 4.1\data\superstore_clean.csv")
print("Saved shape :", check.shape)
print("Cleaned file confirmed ✅")
```
